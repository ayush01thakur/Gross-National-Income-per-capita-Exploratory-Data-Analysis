DATA PREPRATION:
DATA CLEANING AND ARRANGING ACCORDINGLY:

```python
# Drpping the columns which we don't need:
Global_df.drop(['ISO3'], axis=1, inplace=True)
Global_df.head(5)
```

* Checking for missing and null values

```python
print("Number of missing values in columns: \n", Global_df.isnull().sum())
```

* Now we need to manage the NULL values so that we can perform analysis. (replacing NAN with median of respective numeric columns)

```python

for i in Global_df.select_dtypes(include=['float64', 'int64']).columns:
    mid= Global_df[i].median()
    Global_df.fillna(mid, inplace=True)

# printing and checking for null values in columns again
print("Number of missing values in columns: \n", Global_df.isnull().sum())
```

* Checking and removing duplicated data set

```python

Global_df.loc[Global_df.duplicated()]
```

since rows =0 therefore there are no duplicates in data set.

```python
# Now I want to rename the columns in the shorter form i.e. from "Global National Income Per Capita {year}" to GNIP {year}

years= np.arange(1990, 2022)
years=list(years)
newCol= [("GNIPC " + str(x)) for x in years]
col_list= Global_df.columns.to_list()[6:]

k=0
newCol_dict=dict()
for i in range(0, len(col_list)):
    newCol_dict[col_list[i]]=newCol[i]

Global_df.rename(columns=newCol_dict, inplace=True)
Global_df
```

Converting the HDI RANK from Float 64 data type to INT 64 data type because rank can only be an integer

```python
Global_df['HDI Rank (2021)']=Global_df['HDI Rank (2021)'].astype(int)
print("Data type of column HDI Rank (2021): ",Global_df['HDI Rank (2021)'].dtypes)
```

* GROUPING THE CONTINENTS WITH THEIR MEANS OF GNIPC {YEAR}

```python

GroupConti_df=Global_df.groupby('Continent').agg(func='mean',numeric_only=True)
GroupConti_df=GroupConti_df.drop('HDI Rank (2021)',axis=1)
GroupConti_df=GroupConti_df.reset_index()
GroupConti_df.head()
```

* Building a new dataset so that we can plot the line chart on the basis of continental progress based on time period

```python

continents=['Africa', 'America', 'Asia', 'Europe', 'Oceania']
newList= GroupConti_df[GroupConti_df.columns[1:]].values.tolist()

GrossDataContinent=[]
for j in continents:
    for i in range(len(years)):
        GrossDataContinent.append(j)

GrossContinentYears= []
for j in range(5):
    for i in years:
        GrossContinentYears.append(i)

GrossContinentMean= []
for continent in newList:
    for mean in continent:
        GrossContinentMean.append(mean)       


GrossYearContinentMeanData={'continent': GrossDataContinent, 'years': GrossContinentYears, 'Mean': GrossContinentMean}
newGross_df= pd.DataFrame(GrossYearContinentMeanData)
newGross_df
```

* Now based on the newly created data set above, we are going to plot a line chart to visualize the progress in the Gross National Income Per Capita on the basis of time period which are year from 1990 to 2021

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=newGross_df, x="years", y="Mean", hue='continent')
plt.title("Relation of rank with respect to Gross National Income Per capita based on Continent year 1990")
plt.show()
```

* Now want to compare the relationship between the HDI rank and the Gross National Income Per Capita 

```python
# Comparing the year 1990 with year 2021 using scatterplot
plt.figure(figsize=(15,6))
sns.scatterplot(data=Global_df, x='GNIPC 1990', y="HDI Rank (2021)", hue='Continent')
plt.title("Relation of rank with respect to Gross National Income Per capita based on Continent year 1990")
plt.show()
```

```python
# year 2021 

plt.figure(figsize=(15,6))
sns.scatterplot(data=Global_df, x='GNIPC 2021', y="HDI Rank (2021)", hue='Continent')
plt.title("Relation of rank with respect to Gross National Income Per capita based on Continent year 2021")
plt.show()
```

Thus we can say lower the GNIPC, higher the HDI rank. And, in both years Africa is the least at GNIPC so its rank is closer to 200. Which is considerably last in both case. Moreover, Europe has the highest rank which is near to 1st or 1st for Human development index in both the year 1990 and 2021.

* Making box plot to summrize the aggregates like mean median and (25th, 75th) percentiles also the outliers

```python

custom_palette = sns.color_palette("coolwarm", n_colors=len(GroupConti_df))
plt.figure(figsize=(30,10))
sns.boxplot(data=GroupConti_df, palette=custom_palette)

plt.title("Box Plot for Every year's GNIPC")
plt.xlabel("YEARS")
plt.xticks(rotation=45)
plt.ylabel("GROSS NATIONAL INCOME PER CAPITA")
plt.show()
```
